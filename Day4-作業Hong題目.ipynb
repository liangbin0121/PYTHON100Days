題目:<br>
english_score = np.array([55,89,76,65,48,70])<br>
math_score = np.array([60,85,60,68,55,60])<br>
chinese_score = np.array([65,90,82,72,66,77])<br>
上3列共六位同學的英文、數學、國文成績，第一個元素代表第一位同學，舉例第一位同學英文55分、數學60分、國文65分，運用上列數據回答下列問題。<br>
1. 有多少學生英文成績比數學成績高?
2. 是否全班同學最高分都是國文?




```
import numpy as np
english_score = np.array([55,89,76,65,48,70])
math_score = np.array([60,85,60,68,55,60])
chinese_score = np.array([65,90,82,72,66,77])
```

```
#1.有多少學生英文成績比數學成績高?
e_m_compar = np.greater(english_score,math_score)
e_m_compar
cnt =0
for i in e_m_compar:
  if i == True:
    cnt = cnt+1
print(f"{cnt} 位學生英文成績比數學成績高")    
```

```
#2.是否全班同學最高分都是國文?
##english_score/math_score/chinese_score

e_m_compar = np.greater(english_score,math_score)
cnt =0
for i in e_m_compar:
  if i == True:
    cnt = cnt+1
if cnt == 6:
  e_m_compar = np.greater(english_score,chinese_score)
  cnt =0
  for i in e_m_compar:
    if i == True:
      cnt = cnt+1
      if cnt == 6:
        print(f"全班同學最高分都是英文")

e_m_compar = np.greater(math_score,english_score)
cnt =0
for i in e_m_compar:
  if i == True:
    cnt = cnt+1
if cnt == 6:
  e_m_compar = np.greater(math_score,chinese_score)
  cnt =0
  for i in e_m_compar:
    if i == True:
      cnt = cnt+1
      if cnt == 6:
        print(f"全班同學最高分都是數學")

e_m_compar = np.greater(chinese_score,english_score)
cnt =0
for i in e_m_compar:
  if i == True:
    cnt = cnt+1
if cnt == 6:
  e_m_compar = np.greater(chinese_score,math_score)
  cnt =0
  for i in e_m_compar:
    if i == True:
      cnt = cnt+1
      if cnt == 6:
        print(f"全班同學最高分都是國文")
```
