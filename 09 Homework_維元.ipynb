### 2. 請問如果現在有一個 DataFrame 如下，請問資料在 Python 中可能長怎樣？

```
#      city  visitor weekday
# 0  Austin      139     Sun
# 1  Dallas      237     Sun
# 2  Austin      326     Mon
# 3  Dallas      456     Mon
```



````python
df=pd.DataFrame({
    'city':['Austin','Dallas','Austin','Dallas'],
    'visitor':[139,237,326,456],
    'weekday':['Sun','Sun','Mon','Mon']
})
print(df)
df
````

### 3. 承上題，假設你想知道每個 weekday 的平均 visitor 數量，可以怎麼做？


````python
sun_mean=df[df['weekday']=='Sun']['visitor'].mean()
mon_mean=df[df['weekday']=='Mon']['visitor'].mean()
print(sun_mean)
print(mon_mean)
````
