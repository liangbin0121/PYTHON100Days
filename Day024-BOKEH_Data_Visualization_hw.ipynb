```python
import bokeh.io
from bokeh.resources import INLINE
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.sampledata.stocks import AAPL, GOOG, IBM, MSFT
import pandas as pd

# 環境 settings
bokeh.io.reset_output()
bokeh.io.output_notebook(INLINE)

```

```python
#指定輸出檔名
output_file("bars24.html")

#給定資料
fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries'] 
counts = [5, 3, 4, 2, 4, 6]

# 不增設按鈕 toolbar_location=None, tools=""
# plot_height 設定條形高度
#配置圖表
p=figure(x_range=fruits, plot_height=250,title="Fruit Counts",
        toolbar_location=None, tools="") 

p.vbar(x=fruits, top=counts, width=0.9)

#Y座標從0開始
p.y_range.start = 0

show(p)
```

```python
#指定輸出檔名
output_file("fruits_sorted.html")

#給定資料
fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries'] 
counts = [5, 3, 4, 2, 4, 6]

#資料做排序
sorted_fruits = sorted(fruits, key=lambda x: counts[fruits.index(x)])
sorted_counts = sorted(counts,key=lambda x:x)

# 不增設按鈕 toolbar_location=None, tools=""
# plot_height 設定條形高度
# 配置圖表

p=figure(x_range=sorted_fruits, plot_height=250,title="Fruit Counts",
        toolbar_location=None, tools="") 

p.vbar(x=sorted_fruits, top=sorted_counts, width=0.9)

#Y座標從0開始
p.y_range.start = 0

show(p)
```

對 fruits 做順序排序

```python
output_file("only_fruits_sorted.html")

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries'] 
counts = [5, 3, 4, 2, 4, 6]

sorted_fruits = sorted(fruits, key=lambda x: counts[fruits.index(x)])

p=figure(x_range=sorted_fruits, plot_height=250,title="Fruit Counts",
        toolbar_location=None, tools="") 
p.vbar(x=fruits, top=counts, width=0.9)
p.y_range.start=0
show(p)
```
