### 1. 根據題目給的 DataFrame 完成下列操作：

```
- 計算每個不同種類 animal 的 age 的平均數
- 計算每個不同種類 animal 的 age 的平均數
- 將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
- 將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
```


````python
# 記得先 Import 正確的套件
# 載入 NumPy, Pandas 套件
import numpy as np
import pandas as pd

# 檢查正確載入與版本
print(np)
print(np.__version__)
print(pd)
print(pd.__version__)
````

````python
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
````

````python

dog_avg=df[df['animal']=='dog'].age.mean()
print('dog age avg:',dog_avg)
cat_avg=df[df['animal']=='cat'].age.mean()
print('cat age avg:',cat_avg)
snake_avg=df[df['animal']=='snake'].age.mean()
print('snake age avg:',snake_avg)

animal_sort=df.sort_values(by=['age','visits'], ascending=False)
print('animal sort age visits:',animal_sort)
df['priority_TF']=df.priority.apply(lambda x:'True' if x=='yes' else 'False')
print('priority TF:',df)
````

### 2. 一個包含兩個欄位的 DataFrame，將每個數字減去

* 1) 該欄位的平均數
* 2) 該筆資料平均數

````python
df = pd.DataFrame(np.random.random(size=(5, 3)))
df
````

````python
# 1) 該欄位的平均數
# 2) 該筆資料平均數
print(df.mean())
print(df.mean(axis=1))
````
