```
import numpy as np
```

```
#. 將下兩列array存成npz檔
array1 = np.array(range(30))
array2 = np.array([2,3,5])


with open('multi_array_h06.npz', 'wb') as f:
    np.savez(f, array1=array1, array2=array2)
```

```
npzfile = np.load('multi_array_h06.npz')
type(npzfile)
```

```
npzfile.files
```

```
#2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔
z = np.random.rand(10)
array_1=npzfile['array1']
array_2=npzfile['array2']

with open('multi_array_h06.npz', 'wb') as f:
    np.savez(f, array1=array_1, array2=array_2, array3=z)
```

```
npzfile = np.load('multi_array_h06.npz')
type(npzfile)
```
