題目:<br>
1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。<br>
2.呈上題，將以上數列取出偶數。<br>
3.呈1題，將數列取出3的倍數。<br>

```
import numpy as np
```

```
#1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。
a=np.arange(0,21)
a
```

```
#2.呈上題，將以上數列取出偶數。
b=a[a%2==0]
b
```

```
#3.呈1題，將數列取出3的倍數。
c=a[a%3==0]
c
```
