### Q1:計算標準常態分配，小於1的機率有多大?
![](Q1.png)

```python
# library
# library

from numpy.random import normal,uniform
from numpy.random import seed
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
import pylab
%matplotlib inline
```

```python
## Q1
## 常態分布的計算
## 計算標準常態分布 X小於1的比例
mu=0
sigma=1
b=stats.norm.cdf(1,mu,sigma)
print("P(Z<1)=",b)
```

### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
![](Q2.png)


```python
## Q2
## 先算(-1<X<1), 再算P(X>1 and X<-1)
mu=0
sigma=1
b=stats.norm.cdf(1,mu,sigma)
a=stats.norm.cdf(-1,mu,sigma)
print("P(X>1 and X<-1)=",1-(b-a))
```

### Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?

```python

##算法1
mu=2
sigma=2
b=stats.norm.cdf(6,mu,sigma)
print("P(X<6)=",b)
```

```python
## Q3
#  提示: 和 Q1,Q2 不同，要先標準化。
##算法2
mu=0
sigma=1
n_c=(6-2)/(4**(0.5))
b=stats.norm.cdf(n_c,mu,sigma)
print("P(X<2)=",b)
```
