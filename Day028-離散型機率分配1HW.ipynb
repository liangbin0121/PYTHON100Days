### 丟一個銅板，丟了100次，出現正面 50 次的機率有多大。
(提示: 先想是哪一種分配，然後透過 python 語法進行計算)


```python
# library
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
```

```python
p=0.5
n=100
x=np.arange(0,100)
```

```python
prob=stats.binom.pmf(50,n,p)
print(f'100次中出現50次正面的機率為: {prob.round(4)}')
```

```python
probs=stats.binom.pmf(x,n,p)
plt.bar(x,probs)
plt.ylabel("p(X=x)")
plt.xlabel('x')
plt.title('binomial(n=100,p=0.5)')
plt.show()
```
